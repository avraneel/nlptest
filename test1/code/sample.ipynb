```python
import pandas as pd
import gensim
import re
import torch
import fasttext
import nltk
```

```python
df = pd.read_csv('../data/movie_review.csv')
df.head()
```

```python
def preprocess(text):
    text = re.sub(r'[^\s\w]', '', text)
    return text.lower()

df['text'] = df.text.apply(preprocess)
df['text'] = df.text.apply(gensim.utils.simple_preprocess)
df.head()
```

```python
text = df.text
text.head()
print(len(text[0]))
```

```python
text.head()
text.to_csv('../data/movie_reviews_preprocessed.csv', index=False, header=False)
```

```python
model = gensim.models.FastText(workers=8)
model.build_vocab(corpus_iterable=text)
```

```python
model.train(corpus_iterable=text, total_examples=len(text), epochs=10)
```

```python
model.save('../models/fasttext_movies.model')
```
