```python
import pandas as pd, re

df = pd.read_csv('../data/New_Delhi_reviews.csv')
df.dropna(inplace=True)

df.astype(str)
```

```python
def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)

df.review_full.apply(remove_emojis)

def remove_link(data):
    return re.sub(r'http\S+', '', data)

def remove_punctuation(data):
    return re.sub(r'[^\w\s]', '', data)

def lower(data):
    return data.lower()

df['review_full'] = df.review_full.apply(remove_link)
df['review_full'] = df.review_full.apply(remove_punctuation)
df['review_full'] = df.review_full.apply(lower)

review 
print(df.head())
```

```python
df.to_csv('../data/New_Delhi_reviews_rnn.csv', index=False, encoding='utf-8')
```
